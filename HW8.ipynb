```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_excel('1.4_02.xlsx')
df.head(10)
```

```python
df['рік'] = df.apply(lambda x: x['Регіон, рік'][-4:], axis=1)
df['Регіон'] = df.apply(lambda x: x['Регіон, рік'][:-6], axis=1)
df.head(10)
```

```python
new_df = {'year':[], 'region':[], 'type':[], 'percantage':[]}
types = ['сільське господарство','промисловість','будівництво','послуги','державне управління']
for index, row in df.iterrows():
    for i in range(len(types)):
        new_df['year'].append(row['рік'])
        new_df['type'].append(types[i])
        new_df['percantage'].append(row[types[i]])
        new_df['region'].append(row['Регіон'])
new_df = pd.DataFrame(new_df)        
new_df.head()        
```

```python
alt.Chart(new_df, title="The Morris Mistake").mark_point().encode(
    alt.X(
        'percantage:Q',
        title="Barley Yield (bushels/acre)",
        scale=alt.Scale(zero=False),
        axis=alt.Axis(grid=False)
    ),
    alt.Y(
        'type:N',
        title="",
        sort='-x',
        axis=alt.Axis(grid=True)
    ),
    color=alt.Color('year:N', legend=alt.Legend(title="Year")),
    row=alt.Row(
        'region:N',
        title="",
        sort=alt.EncodingSortField(field='yield', op='sum', order='descending'),
    )
).properties(
    height=alt.Step(20)
).configure_view()
```
